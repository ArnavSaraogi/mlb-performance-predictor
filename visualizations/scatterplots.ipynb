# Scatterplots

```python
import os
import pandas as pd

df = pd.read_csv('../out.csv')
```

```python
ax1 = (df.plot(title = "2021-2022 %ΔOPS vs 2021 wOBAdiff", 
                kind = 'scatter', 
                x = 'wobadiff (2021)', 
                y = 'ops_change 2021-2022', 
                xlabel = 'wOBA-xwOBA',
                ylabel = '%ΔOPS'))

ax2 = (df.plot(title = "2022-2023 %ΔOPS vs 2022 wOBAdiff", 
                kind = 'scatter', 
                x = 'wobadiff (2022)', 
                y = 'ops_change 2022-2023', 
                xlabel = 'wOBA-xwOBA',
                ylabel = '%ΔOPS'))

ax3 = (df.plot(title = "2023-2024 %ΔOPS vs 2023 wOBAdiff", 
                kind = 'scatter', 
                x = 'wobadiff (2023)', 
                y = 'ops_change 2023-2024', 
                xlabel = 'wOBA-xwOBA',
                ylabel = '%ΔOPS'))
```
